### 1.年度2010~2020年資料

```python
df=df.loc['2010':'2020']
df
```

### 2.波動度計算

```python
df['Return']=(df['Adj Close']-df['Adj Close'].shift(1))/df['Adj Close'].shift(1)
df
```

### 3.儲存csv檔案

```python
df.to_csv('yfinance.csv',encoding='utf-8-sig')
```

### 呼叫csv檔案
- 套用pandas
- 呼叫csv
- 轉index ==> .to_datetime()

```python
df1=pd.read_csv('ch15_test001.csv',index_col=0)
df1.index=pd.to_datetime(df1.index)
df1
```

### 4.繪製日報酬波動圖

```python
df1['Return'].plot(figsize=(12,6))
```

#### 找出最高與最低,日期

```python
f'最低點日期:{df1["Return"].idxmax()} 最低點為:{df1["Return"].max()}'
```

```python
f'最低點日期:{df1["Return"].idxmin()} 最低點為:{df1["Return"].min()}'
```

### 6.年報酬率(柱狀圖)

```python
df1_year=df1.resample('AS').sum()['Return']
df1_year
```
